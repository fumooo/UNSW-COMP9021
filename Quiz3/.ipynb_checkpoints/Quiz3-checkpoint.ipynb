```python
# Written by *** for COMP9021
#
# Implements a function that takes as argument a string
# consisting of arrows pointing North, East, South or West.
#
# Following the provided directions,
# - if the exploration gets back to the starting point, then this
#   common location will be represented by a black circle;
# - otherwise, the starting point will be represented by a blue circle
#   and the final destination by a red circle.

# All other visited locations will be represented by a square, of colour:
# - yellow if visited exactly once, 6 times, 11 times, 16 times...
# - orange if visited exactly twice, 7 times, 12 times, 17 times...
# - brown if visited exactly trice, 8 times, 13 times, 18 times...
# - green if visited exactly 4 times, 9 times, 14 times, 19 times...
# - purple if visited exactly 5 times, 10 times, 15 times, 20 times...

# The explored area is displayed within the smallest rectangle in
# which it fits; all unvisited locations within that rectangle are
# represented by white squares.
#
# The code points of the characters involved in this quiz are:
# 9899, 11036, 128308, 128309, 128999, 129000, 129001, 129002, 129003

#     print("â¬†" + chr(11014))
#     print("â®•" + chr(11157))
#     print("â¬‡" + chr(11015))
#     print("â¬…" + chr(11013))
#     print("9899" + chr(9899))
#     print("11036" + chr(11036))
#     print("128308" + chr(128308))
#     print("128309" + chr(128309))
#     print("128999" + chr(128999))
#     print("129000" + chr(129000))
#     print("129001" + chr(129001))
#     print("129002" + chr(129002))
#     print("129003" + chr(129003))
def findMax(Num_Max, tmp):
    if Num_Max < tmp:
        Num_Max = tmp
    return Num_Max
def findMin(Num_Min, tmp):
    if Num_Min > tmp:
        Num_Min = tmp
    return Num_Min
def findColourByVisitTime(VisitTime):
    if(VisitTime == 0):
        return 11036
    if((VisitTime-1)%5 == 0):
        return 129000
    if((VisitTime-2)%5 == 0):
        return 128999
    if((VisitTime-3)%5 == 0):
        return 129003
    if((VisitTime-4)%5 == 0):
        return 129001
    if((VisitTime)%5 == 0):
        return 129002
def explore_this_way(directions):
    directions_list = list(directions)
    height = 0
    height_max = 0
    height_min = 0
    length = 0
    length_max = 0
    length_min = 0
    
    start_P_Length = 0 #the start point position
    start_P_Height = 0 #the start point position
    for i in directions_list:
        if(i == "â¬†"):
            height = height + 1
            if(height_max < findMax(height_max, height)):
                start_P_Height = start_P_Height + 1
            height_max = findMax(height_max, height)
        if(i == "â¬‡"):
            height = height - 1
            height_min = findMin(height_min, height)
        if(i == "â®•"):
            length = length + 1
            length_max = findMax(length_max, length)
        if(i == "â¬…"):
            length = length - 1
            if(length_min > findMin(length_min, length)):
                start_P_Length = start_P_Length + 1
            length_min = findMin(length_min, length)
    print("This is the start point position:" , start_P_Height , start_P_Length)
    list_height = height_max - height_min + 1
    list_length = length_max - length_min + 1

    Matrix = [[0 for i in range(list_length)] for j in range(list_height)]
    current_point_length = start_P_Length # the current position
    current_point_height = start_P_Height # the current position
    for i in directions_list:
        if(i == "â¬†"):
            current_point_height = current_point_height - 1
            Matrix[current_point_height][current_point_length] = Matrix[current_point_height][current_point_length] + 1
        if(i == "â¬‡"):
            current_point_height = current_point_height + 1
            Matrix[current_point_height][current_point_length] = Matrix[current_point_height][current_point_length] + 1
        if(i == "â®•"):
            current_point_length = current_point_length + 1 
            Matrix[current_point_height][current_point_length] = Matrix[current_point_height][current_point_length] + 1
        if(i == "â¬…"):
            current_point_length = current_point_length - 1
            Matrix[current_point_height][current_point_length] = Matrix[current_point_height][current_point_length] + 1
    print("This is the ending point:",current_point_height,current_point_length)
    for i in range(len(Matrix)):
        for m in range(len(Matrix[0])):
            Matrix[i][m] = findColourByVisitTime(m)

    if(start_P_Height == current_point_height and start_P_Length == current_point_length):
        Matrix[start_P_Height][start_P_Length] = 9899
    else: 
        Matrix[start_P_Height][start_P_Length] = 128309
        Matrix[current_point_height][current_point_length] = 128308
    
    cal = 0   
    for i in Matrix:
        for m in i:
            print(chr(m))
            cal = cal + 1
            if(cal % list_length == 0):
                print()
    # REPLACE PASS ABOVE WITH YOUR CODE

explore_this_way('â¬†' * 4 + 'â®•' * 4 + 'â¬‡' * 1 + 'â¬…' * 5)
```
